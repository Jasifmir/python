```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data=[[2,2,10,20,3],[20,13,11,7,5],[2,2,7,19,15]]
x=np.arange(5)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x + 0.00, data[0], color='r', width = 0.5)
ax.bar(x + 0.25, data[1], color='g', width = 0.15)
ax.bar(x + 0.50, data[2], color='b', width = 0.10)
plt.show()
```
